```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1. Bar chart Response time

```python
Algorithms =['ArraySum','BubbleSort','NumberCheck','ContainsSubstring','Power','BinaryConversion','MatrixMultiply','PrimeNumber','Total']
WOC=[41,91,65,75,66,76,83,51,69]
WIC = [64,115,51,91,68,52,84,83,76]
CNL = [64,60,74,92,48,86,113,53,74]
```

```python
xpos=np.arange(len(Algorithms))
```

```python
xpos
```

```python
plt.xticks(xpos,Algorithms,rotation=60)
plt.xlabel('Algorithms')
plt.ylabel('ResponseTime(Sec.)')
plt.bar(xpos+0.1,WOC, label='WOC',width=0.2,align='center')
plt.bar(xpos+0.3,WIC,label='WIC',width=0.2,align='center')
plt.bar(xpos+0.5,CNL,label='CNL',width=0.2,align='center')
plt.legend()
resolution=1200
plt.savefig('responsetime.pdf',bbox_inches='tight',dpi=resolution)
```
